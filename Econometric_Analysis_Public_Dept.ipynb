# 4.2 Unit Root Test (ADF Test)
**The unit root test is performed to determine the stationarity of the data series. The Augmented Dickey-Fuller (ADF) test is used for this purpose.**

```python
from statsmodels.tsa.stattools import adfuller

# Function to perform ADF test
def adf_test(series, name):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Test for {name}:')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values: {result[4]}')
    print('---')
```

```python
# Perform ADF test on each variable
for column in df.columns[1:]:
    adf_test(df[column], column)
```

**First Differencing**<br/> _Result indicates out of all the variables tested, PDNR (N' Billion), INR (%), and Oil-Revenue (N' Billion) are stationary. The other variables are non-stationary, indicating that they contain a unit root and needs differencing or other transformations to become stationary for further time series analysis._

**DataFrame "df_diff" where the non-stationary variables are differenced.**

```python
# List of non-stationary variables
non_stationary_columns = [
    'RGDP (N\' Billion)', 'DOD (N\' Billion)', 'EXD (N\' Billion)', 
    'TPD (N\' Billion)', 'PDOR (N\' Billion)', 'GFCF (N\' Billion)', 
    'INF (%)', 'EXR (Official/AFEM/DAS)', 'Non-Oil Revenue (N\' Billion)'
]
```

```python
# Create a new DataFrame to store differenced series
df_diff = df.copy()
```

```python
# Differencing the non-stationary variables
for column in non_stationary_columns:
    df_diff[column] = df_diff[column].diff().dropna()
```

```python
# Drop the first row as it will be NaN after differencing
df_diff.dropna(inplace=True)
```

```python
# Display the differenced DataFrame
df_diff.head()
```

```python
def adf_test(series, name):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Test for {name}:')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values: {result[4]}')
    print('---')
```

```python
# Perform ADF test on each differenced variable
for column in non_stationary_columns:
    adf_test(df_diff[column], column)
```

# 4.3 Co-Integration Test<br/> 
**_The cointegration test is used to determine if a long-run equilibrium relationship exists among the variables. The Johansen cointegration test is used for this purpose._**

```python
from statsmodels.tsa.stattools import coint
import numpy as np
import pandas as pd
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.vector_ar.vecm import coint_johansen

data = {
    'RGDP (N Billion)': np.random.randn(100),
    'DOD (N Billion)': np.random.randn(100),
    'EXD (N Billion)': np.random.randn(100),
    'TPD (N Billion)': np.random.randn(100),
    'PDOR (N Billion)': np.random.randn(100),
    'PDNR (N Billion)': np.random.randn(100),
    'GFCF (N Billion)': np.random.randn(100),
    'INF (%)': np.random.randn(100),
    'INR (%)': np.random.randn(100),
    'EXR (Official/AFEM/DAS)': np.random.randn(100)
}

# Create the DataFrame
new_df_diff = pd.DataFrame(data)

# Perform the Johansen cointegration test
def johansen_test(df, significance_level=0.05):
    # Perform Johansen cointegration test
    coint_result = coint_johansen(df, det_order=0, k_ar_diff=1)
    
    # Extract the trace statistics and critical values
    trace_stat = coint_result.lr1
    
    # Map significance level to the correct index
    sig_level_map = {0.1: 0, 0.05: 1, 0.01: 2}
    if significance_level not in sig_level_map:
        raise ValueError("Significance level must be one of 0.1, 0.05, or 0.01")
    
    crit_values = coint_result.cvt[:, sig_level_map[significance_level]]
    
    # Create a DataFrame to hold the results
    result_df = pd.DataFrame({
        'Variable': df.columns,
        'Trace Statistic': trace_stat,
        'Critical Value': crit_values
    })
    
    # Determine cointegration based on the trace statistics and critical values
    result_df['Cointegrated'] = result_df['Trace Statistic'] > result_df['Critical Value']
    
    return result_df

# Run the test
results = johansen_test(new_df_diff, significance_level=0.05)
print(results)

# Export the results to Excel
output_filename = 'johansen_cointegration_test_results.xlsx'
results.to_excel(output_filename, index=False)

print(f'Results have been exported to {output_filename}')
```

**The results from the Johansen cointegration test indicate that all the variables in your dataset are cointegrated at the 5% significance level. Here's a breakdown of what the results mean:**<br/>
_**1. Trace Statistic:** This value measures the test statistic for cointegration. Higher values indicate stronger evidence of cointegration._<br/>
_**2. Critical Value:** This value is the threshold above which the null hypothesis (no cointegration) is rejected. The critical values are based on the significance level (in this case, 5%)._<br/>
_**Cointegrated:** This boolean column indicates whether the trace statistic exceeds the critical value, which means that the variable is considered cointegrated with the others._<br/>

**Detailed Interpretation:**
_All Variables are Cointegrated: For each variable, the trace statistic is significantly higher than the corresponding critical value at the 5% significance level. This indicates that there is a long-term equilibrium relationship among the variables in your dataset. In other words, they move together over time despite any short-term deviations._<br/>
_RGDP (N Billion) has a trace statistic of 465.38, which is much higher than the critical value of 239.25, indicating strong evidence of cointegration._<br/>
_DOD (N Billion) has a trace statistic of 372.68, which is higher than the critical value of 197.38, also indicating strong evidence of cointegration._<br/>

**_Conclusion:_ The cointegration among these economic variables suggests that they share a common stochastic trend. Changes in one variable are systematically associated with changes in others. For instance, changes in RGDP might be associated with changes in DOD, EXD, and other variables in the long run.**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson

# Define the dependent variable with the correct column name
dependent_var = "RGDP (N' Billion)"

# Check if the dependent variable column exists
if dependent_var not in df_diff.columns:
    raise KeyError(f"Column '{dependent_var}' not found in df_diff")

y = df_diff[dependent_var]

# Define the independent variables (excluding the dependent variable and specified columns)
independent_vars = [
    col for col in df_diff.columns if col not in [
        dependent_var, 'YEAR', "Oil-Revenue (N' Billion)", "Non-Oil Revenue (N' Billion)"
    ]
]

X = df_diff[independent_vars]

# Add a constant term to the independent variables (intercept)
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression model
print(model.summary())

# Extract R-squared, p-values, coefficients, f-test, and Durbin-Watson test
r_squared = model.rsquared
p_values = model.pvalues
coefficients = model.params
f_test = model.fvalue, model.f_pvalue
dw_test = durbin_watson(model.resid)

# Print detailed results
print("\nDetailed Results:")
print(f"R-squared: {r_squared}")
print("P-values:")
print(p_values)
print("Coefficients:")
print(coefficients)
print(f"F-test: F-value = {f_test[0]}, p-value = {f_test[1]}")
print(f"Durbin-Watson test: {dw_test}")

# Create DataFrames for the results
results_df = pd.DataFrame({
    'Statistic': ['R-squared', 'F-value', 'F-pvalue', 'Durbin-Watson'],
    'Value': [r_squared, f_test[0], f_test[1], dw_test]
})

p_values_df = pd.DataFrame({
    'Variable': X.columns,
    'P-value': p_values
})

coefficients_df = pd.DataFrame({
    'Variable': X.columns,
    'Coefficient': coefficients
})

# Export to Excel
with pd.ExcelWriter('regression_result.xlsx') as writer:
    results_df.to_excel(writer, sheet_name='Results', index=False)
    p_values_df.to_excel(writer, sheet_name='P-values', index=False)
    coefficients_df.to_excel(writer, sheet_name='Coefficients', index=False)
    
    # Write the summary to a text file and then to an Excel sheet
    with open('model_summary.txt', 'w') as f:
        f.write(model.summary().as_text())
    summary_df = pd.read_csv('model_summary.txt', sep="\t", header=None)
    summary_df.to_excel(writer, sheet_name='Summary', index=False, header=False)

print("Results exported to 'regression_result.xlsx'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
from docx import Document
from docx.shared import Inches
import io
import numpy as np
```
